Q-3:WAPP to implement a Double bar graph

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values1 = [3, 8, 1, 10]
values2 = [4, 7, 2, 9]


# Set the width of the bars
bar_width = 0.35

# Set the positions of the bars on the x-axis
r1 = range(len(categories))
r2 = [x + bar_width for x in r1]

# Create the bars
plt.bar(r1, values1, color='g', width=bar_width, edgecolor='grey', label='Group 1')
plt.bar(r2, values2, color='r', width=bar_width, edgecolor='grey', label='Group 2')


# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Double Bar Graph')
plt.xticks([r + bar_width/2 for r in range(len(categories))], categories)
# Add legend
plt.legend()

# Show the plot
plt.show()
```
