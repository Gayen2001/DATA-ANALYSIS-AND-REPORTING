Q-6:WAPP to implement a line  bar graph

```python
import matplotlib.pyplot as plt

# Sample data for the bar graph
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Sample data for the line graph
line_values = [4, 6, 2, 7, 5]

# Creating the bar graph
plt.bar(categories, values, color='skyblue', label='Bar Values')

# Adding the line graph
plt.plot(categories, line_values, color='red', marker='o', label='Line Values')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph with Line')
plt.legend()

# Displaying the graph
plt.show()
```
