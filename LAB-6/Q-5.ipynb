Q-5:WAPP to implement a stack  bar graph

```python

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C']
values1 = [10, 20, 30]
values2 = [15, 25, 35]
values3 = [20, 30, 40]

# Plotting
plt.bar(categories, values1, label='Values 1')
plt.bar(categories, values2, bottom=values1, label='Values 2')
plt.bar(categories, values3, bottom=[i+j for i,j in zip(values1, values2)], label='Values 3')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Graph ')
plt.legend()

# Display the graph
plt.show()
```
