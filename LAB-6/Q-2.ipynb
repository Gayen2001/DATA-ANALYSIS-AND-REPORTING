Q-2:WAPP to implement a vertical bar graph

```python
import matplotlib.pyplot as plt
```

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 1, 10]

# Create horizontal bar graph
plt.bar(categories, values)

# Add labels and title
plt.ylabel('Values')
plt.xlabel('Categories')
plt.title('Vertical Bar Graph ')

# Show the plot
plt.show()
```
