Q-4:write a python program to draw a multiple area diagram

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


# Data
x = np.arange(1, 11)
y1 = np.random.randint(1, 10, size=10)
y2 = np.random.randint(1, 10, size=10)
y3 = np.random.randint(1, 10, size=10)

# Stack the data
y = np.vstack([y1, y2, y3])

# Plot
fig, ax = plt.subplots()
ax.stackplot(x, y, labels=['Series 1', 'Series 2', 'Series 3'], alpha=0.8)

# Add legend and labels
ax.legend(loc='upper left')
ax.set_title('Multiple Area Diagram')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show plot
plt.show()
```
